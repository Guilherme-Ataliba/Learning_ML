# Any Regression Algorithm

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 100
X = 6 * np.random.rand(m, 1) - 3
# y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1)
y = 0.5*X**2 + np.exp(X) - 2*np.sin(X)*X + 5 + np.random.randn(m, 1)
```

```python
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([-3, 3, 0, 10])
plt.show()
```

# Função Qualquer

```python
X_pol = np.c_[X**2, np.exp(X), X*np.sin(X), np.ones((100, 1))]
```

### Calculando pela Equação Normal

```python
X_b = X_pol

theta_best = np.linalg.inv( X_b.T.dot(X_b) ).dot( X_b.T.dot(y) )
theta_best
```

```python
X_plot = np.linspace(-3, 3, 500)
y_plot = theta_best[0]*X_plot**2 + theta_best[1]*np.exp(X_plot) + theta_best[2]*X_plot*np.sin(X_plot) + theta_best[3]
```

```python
plt.plot(X, y, ".")
plt.plot(X_plot, y_plot)
plt.axis([-3, 3, 0, 10])
plt.show()
```
